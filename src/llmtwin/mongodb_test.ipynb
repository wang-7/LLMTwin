```python
from pymongo import MongoClient
uri = "mongodb://localhost:27017/"
client = MongoClient(uri, connect=True)
```

```python
with client.list_databases() as cursor:
    for database in cursor:
        print(database)
```

```python
collection = client["mongodbVSCodePlaygroundDB"]
with collection.list_collections() as cursor:
    for coll in cursor:
        print(coll)
```
