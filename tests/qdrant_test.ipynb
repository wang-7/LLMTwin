```python
from llmtwin.vector import VectorBaseDocument

class TestDocument(VectorBaseDocument):
    embedding: list[float] | None
    city: str

    @classmethod
    def get_collection_name(cls):
        return 'test'
```

```python
cities = ['Changsha', 'Nanchang']
documents = [TestDocument(city=city, embedding=[1., 2., 3., 4.]) for city in cities]

TestDocument.bulk_insert(documents)
```

```python
docs, next_offset = TestDocument.bulk_find(limit=3)
print('offset:', next_offset)
for doc in docs:
    print(doc)
```

```python
results = TestDocument.query(
    query_vector=[1., 2., 3., 4.],
    limit=2,
)
for result in results:
    print(result)
```
